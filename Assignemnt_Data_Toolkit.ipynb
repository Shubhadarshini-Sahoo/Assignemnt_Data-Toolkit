**[1]What is NumPy, and why is it widely used in Python?**

NumPy (Numerical Python) is a Python library for efficient numerical computing. It provides fast, memory-efficient multidimensional arrays, supports vectorized operations and broadcasting, and offers a wide range of mathematical, statistical, and linear algebra functions. Its speed, ease of use, and integration with libraries like Pandas, SciPy, and Matplotlib make it essential for data analysis, machine learning, and scientific computing.

**[2] How does broadcasting work in NumPy?**

Broadcasting in NumPy is a mechanism that allows arithmetic operations to be performed on arrays of different shapes and sizes. Instead of requiring arrays to have identical shapes for element-wise operations, broadcasting automatically "stretches" or "broadcasts" the smaller array to match the shape of the larger one, making them compatible for the operation. This process happens virtually without creating explicit copies of the data, which enhances performance and memory efficiency.

**[3]What is a Pandas DataFrame?**

A Pandas DataFrame is a two-dimensional, size-mutable, and labeled data structure in Python, similar to a table in a database or an Excel spreadsheet. It allows storing, accessing, and manipulating structured data with rows and columns, supports heterogeneous data types, and provides powerful tools for filtering, aggregation, and analysis.

**[4]Explain the use of the groupby() method in Pandas.**

In Pandas, the groupby() method is used to split data into groups, apply a function, and then combine the results.

```python
import pandas as pd

data = {
    "Department": ["HR", "IT", "HR", "IT", "Finance"],
    "Salary": [3000, 4000, 3500, 4500, 5000]
}
df = pd.DataFrame(data)

# Group by Department and get average salary
result = df.groupby("Department")["Salary"].mean()
print(result)
```

**[6]What are the differences between NumPy arrays and Python lists?**

NumPy arrays and Python lists are both used to store collections of elements, but they differ significantly. NumPy arrays are homogeneous, meaning all elements must be of the same data type, while Python lists can hold mixed types. Arrays are stored in contiguous memory, making them more memory-efficient and enabling faster numerical computations through vectorized operations, unlike lists which require explicit loops. NumPy also provides advanced mathematical, statistical, and linear algebra functions and supports multi-dimensional arrays with powerful indexing, whereas lists are primarily one-dimensional with basic indexing. In short, NumPy arrays are optimized for efficient numerical operations, while lists offer more general-purpose flexibility.

**[8]What does the term “vectorized operation” mean in NumPy?**

In NumPy, a vectorized operation means performing element-wise operations on entire arrays without using explicit Python loops. NumPy uses optimized C code under the hood, so these operations are much faster and more efficient than looping through elements manually.

```python
import numpy as np

a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])

# Vectorized addition (element-wise)
c = a + b
print(c)
```

**[17]Explain the difference between apply() and map() in Pandas.**

In Pandas, map() is used only with a Series to apply a function, dictionary, or mapping element-wise, making it ideal for simple element-level transformations. In contrast, apply() works with both Series and DataFrames; on a Series it behaves like map(), while on a DataFrame it can apply a function across either rows or columns, making it more flexible for complex operations. In short, map() is limited to element-wise operations on Series, whereas apply() is more versatile, supporting both element-wise and axis-wise operations.

```python
import pandas as pd
import numpy as np

s = pd.Series([1, 2, 3, 4])
print(s.map(lambda x: x**2))   # map() → element-wise
print(s.apply(lambda x: x**2)) # apply() on Series

df = pd.DataFrame({"A": [1, 2], "B": [10, 20]})
print(df.apply(np.sum, axis=0))
```

**[20]What is the role of a pivot table in Pandas?**

In Pandas, a pivot table is used to summarize, organize, and analyze data by rearranging it into a tabular format based on one or more keys. It lets you group data, apply aggregations (like sum, mean, count), and compare values across categories easily, similar to Excel’s pivot tables.

```python
import pandas as pd

data = {
    "Department": ["HR", "IT", "HR", "Finance", "IT"],
    "Employee": ["A", "B", "C", "D", "E"],
    "Salary": [3000, 4000, 3500, 5000, 4500]
}

df = pd.DataFrame(data)

pivot = df.pivot_table(values="Salary", index="Department", aggfunc="mean")
print(pivot)
```

**[21]Why is NumPy’s array slicing faster than Python’s list slicing?**

NumPy’s array slicing is faster than Python’s list slicing because of how data is stored and accessed internally:

Contiguous memory → NumPy arrays are stored in a single continuous memory block, so slicing only creates a view (no data is copied), while Python lists are collections of pointers, so slicing requires creating a new list and copying elements.

Typed data → NumPy arrays hold elements of the same type (homogeneous), enabling optimized low-level (C-based) operations, whereas Python lists can hold mixed types, adding overhead.

Vectorized operations → NumPy can process slices with efficient, compiled routines instead of looping through Python objects.

**4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6]
}

df = pd.DataFrame(data)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Visualize with a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

**13.Create a 3D scatter plot using Plotly.**

```python
import plotly.express as px
import pandas as pd

# Sample data
data = {
    'X': [1, 2, 3, 4, 5],
    'Y': [10, 11, 12, 13, 14],
    'Z': [5, 4, 3, 2, 1],
    'Category': ['A', 'B', 'A', 'B', 'A']  # optional for coloring
}

df = pd.DataFrame(data)

# Create 3D scatter plot
fig = px.scatter_3d(df, x='X', y='Y', z='Z', color='Category', size='X', symbol='Category')

# Add title
fig.update_layout(title='3D Scatter Plot Example')

# Show plot
fig.show()
```
